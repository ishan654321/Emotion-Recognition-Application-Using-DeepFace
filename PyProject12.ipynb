```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data and drop NaN values
data = pd.read_csv("/content/India_budget_2021.csv")
data.dropna()

# Select specific rows for visualization
data = data.iloc[[0, 8, 11, 14, 18, 23, 41, 42, 43], :]

# Add 'OTHERS' category
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.08}
data = data.append(row, ignore_index=True)

# Sort data by 'Fund allotted' in descending order
data = data.sort_values(by='Fund allotted(in ₹crores)', ascending=False)

# Format 'Fund allotted' column with commas for better readability
data['Fund allotted(in ₹crores)'] = data['Fund allotted(in ₹crores)'].apply(lambda x: "{:,.2f}".format(x))

# Bar chart
plt.figure(figsize=(12, 6))
plt.bar(data['Department /Ministry'], data['Fund allotted(in ₹crores)'], color='skyblue')
plt.xlabel('Department / Ministry')
plt.ylabel('Fund allotted (in ₹crores)')
plt.title('India Budget 2021 - Department-wise Allocation')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Pie chart
plt.figure(figsize=(7, 7))
plt.pie(data['Fund allotted(in ₹crores)'], labels=data['Department /Ministry'], autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow=True, colors=plt.cm.Paired(np.linspace(0, 1, len(data))))
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.title("Distribution of The Budget", fontsize=20)
plt.legend(data['Department /Ministry'], loc='upper left', bbox_to_anchor=(-0.1, 1.), fontsize=10)
plt.show()
```
